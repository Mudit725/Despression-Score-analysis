```python
#Tokenization
nltk.download('punkt')
#from nltk.tokenize import word_tokenize
from nltk import RegexpTokenizer
tokenizer = RegexpTokenizer(r'w+')
import nltk
from nltk.tokenize import word_tokenize
train_data['user_id'][0] = 1234
train_data['clean_text'][0] = "I like to be in heaven"
train_data['count'][0] = 1
train_data['clean_token'] = train_data['clean_text'].apply(lambda x: word_tokenize(x))
```

```python
#lines_list = tokenize.sent_tokenize(train_data['clean_text'][i])
#lines_list
```

```python
scores = []
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
sem = []
for i in range(len(train_data)):
        #semantic score
    t = train_data['clean_token'][i]
    h_b, h_n = 0, 0 # hit score
    for token in t:
        if token in broad_terms:
            h_b = h_b + 1
        elif token in narrow_terms:
            h_n = h_n + 1
        else:
            continue
    alpha = 0.8 #0.7 0.6
    semantic_score = (alpha*h_b) + ((1-alpha)*h_n)
    sem.append(semantic_score)

sem = np.asarray(sem)
sem = sem.reshape(-1,1)
sem_scaled = min_max_scaler.fit_transform(sem)

for i in range(len(train_data)):
  # Polarity
    sentences = []
    lines_list = tokenize.sent_tokenize(str(train_data['clean_text'][i]))
    sentences.extend(lines_list)
    polarity = 0
    for sentence in sentences:
        sid = SentimentIntensityAnalyzer()
        print(sentence)
        ss = sid.polarity_scores(sentence)
        # for k in sorted(ss):
        #     print('{0}: {1}, '.format(k, ss[k]), end='')
        # print()
        polarity = polarity + ss['compound']
        polarity *= -1
    dep_score = sem_scaled[i] + polarity
    scores.append([polarity, sem_scaled[i], dep_score])
```

```python
train_data['polarity'] = 0.0
train_data['semantic_score'] = 0.0
train_data['dep_score'] = 0.0
for i in range(len(train_data)):
  train_data['polarity'][i] = scores[i][0]
  train_data['semantic_score'][i] = scores[i][1]
  train_data['dep_score'][i] = scores[i][2]
```
